### Boxplot MEMOTE Score Values

```python
import pandas as pd
from matplotlib import rcParams
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read in the meta-data file as a data frame
meta_df = pd.read_csv('IAMM_models/strain_metadata.tsv', sep='\t')
```

```python
# Create a figure instance
fig = plt.figure()

# Create an axes instance
ax = fig.add_axes([0,0,1,1])

# Create the boxplot
bp = ax.violinplot(meta_df['memote_score'], showmedians = True)

# Color of boxplot
bp['bodies'][0].set_facecolor('#004263')
bp['cbars'].set_facecolor('#004263')
# Define a larger "minimal" media
# Add the individual points
for i, d in enumerate(meta_df['memote_score']):
   y = meta_df['memote_score'][i]
   x = np.random.normal(1, 0.04)
   plt.scatter(x, y, s = 150)

# Style
plt.ylim([0.5, 1])

rcParams.update({'font.size': 40})

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.spines['left'].set_color('#818689')
ax.yaxis.label.set_color('#818689')
ax.tick_params(axis='y', colors='#818689')

ax.spines['bottom'].set_color('#818689')
ax.get_xaxis().set_ticks([])

# Labels
plt.ylabel('MEMOTE Total Score')

# Show
fig.set_size_inches(18.5, 10.5)
plt.show()
```
